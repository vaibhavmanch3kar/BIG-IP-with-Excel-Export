```python
import pandas as pd
```

```python
df = pd.read_csv('wine.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
'''
Year :
Price: Wine Price
WinterRain : winter rainfall
HarvestRain : Harvest Rain
AGST : Average Growing season Temperature
Age: age of wine
FrancePop : population of France
'''

```

```python
'''
Read the data

EDA (Exploratory data analysis)

Preprocessing (Data cleaning)

Choose X and y

Split the data in train and test

Build the model on train data
  Prediction on train data
  Evaluation on train data

  Prediction on test data
  Evaluation on test data

'''
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.columns
```

```python
# Choose X and y

X = df[['AGST']]
y = df[['Price']]
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
# Split the data in train and test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=123)
X_test
```

```python
X_train.head()
```

```python
print("X_train shape : ",X_train.shape)
print("X_test shape : ",X_test.shape)
print("y_train shape : ",y_train.shape)
print("y_test shape : ",y_test.shape)
```

```python
# Build the model on train data

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train) # model training
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# y = b x +C
# Price = 0.74219244 * AGST -5.13135001
```

```python
prediction_train = lr.predict(X_train)
prediction_train
```

```python
# Model Evalaution on train data
from sklearn import metrics
```

```python
# Actual --> y_train
# Prediction --> prediction_train

mse_train = metrics.mean_squared_error(y_train, prediction_train)
print("Mean Squared Error for train data : ", round(mse_train,3))

rmse_train = metrics.root_mean_squared_error(y_train, prediction_train)
print("Root Mean Squared Error for train data : ", round(rmse_train,3))

mae_train = metrics.mean_absolute_error(y_train, prediction_train)
print("Mean Absolute Error for train data : ", round(mae_train,3))

mape_train = metrics.mean_absolute_percentage_error(y_train, prediction_train)
print("Mean Absolute Percentage Error for train data : ", round(mape_train,3))
```

```python
prediction_test = lr.predict(X_test)
prediction_test
```

```python
# Actual --> y_test
# Prediction --> prediction_test

mse_test = metrics.mean_squared_error(y_test, prediction_test)
print("Mean Squared Error for test data : ", round(mse_test,3))

rmse_test = metrics.root_mean_squared_error(y_test, prediction_test)
print("Root Mean Squared Error for test data : ", round(rmse_test,3))

mae_test = metrics.mean_absolute_error(y_test, prediction_test)
print("Mean Absolute Error for test data : ", round(mae_test,3))

mape_test = metrics.mean_absolute_percentage_error(y_test, prediction_test)
print("Mean Absolute Percentage Error for test data : ", round(mape_test,3))
```
